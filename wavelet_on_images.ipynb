```python
import cv2
import numpy as np
import pywt
import matplotlib.pyplot as plt
from scipy.ndimage import center_of_mass

# Function to calculate the azimuthal average
def azimuthal_average(image, center=None):
    y, x = np.indices(image.shape)

    if not center:
        center = center_of_mass(image)

    r = np.hypot(x - center[1], y - center[0])

    r = r.astype(int)  # Use int instead of np.int
    tbin = np.bincount(r.ravel(), image.ravel())
    nr = np.bincount(r.ravel())

    # Avoid division by zero
    radialprofile = np.divide(tbin, nr, out=np.zeros_like(tbin, dtype=float), where=nr != 0)
    return radialprofile

# Load an image
image = cv2.imread('/content/drive/MyDrive/Colab Notebooks/soft_comp/fake/03_11__exit_phone_room__P08VGHTA/15900.jpg', cv2.IMREAD_GRAYSCALE)

# Perform wavelet transform
coeffs2 = pywt.dwt2(image, 'haar')
LL, (LH, HL, HH) = coeffs2

# Calculate azimuthal averages
azimuthal_LL = azimuthal_average(LL)
azimuthal_LH = azimuthal_average(LH)
azimuthal_HL = azimuthal_average(HL)
azimuthal_HH = azimuthal_average(HH)

# Plot the azimuthal averages
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
fig.suptitle('Azimuthal Averages of Wavelet Transform Components')

axes[0, 0].plot(azimuthal_LL)
axes[0, 0].set_title('Azimuthal Average of Approximation (LL)')
axes[0, 0].set_xlabel('Radius')
axes[0, 0].set_ylabel('Average Intensity')

axes[0, 1].plot(azimuthal_LH)
axes[0, 1].set_title('Azimuthal Average of Horizontal Detail (LH)')
axes[0, 1].set_xlabel('Radius')
axes[0, 1].set_ylabel('Average Intensity')

axes[1, 0].plot(azimuthal_HL)
axes[1, 0].set_title('Azimuthal Average of Vertical Detail (HL)')
axes[1, 0].set_xlabel('Radius')
axes[1, 0].set_ylabel('Average Intensity')

axes[1, 1].plot(azimuthal_HH)
axes[1, 1].set_title('Azimuthal Average of Diagonal Detail (HH)')
axes[1, 1].set_xlabel('Radius')
axes[1, 1].set_ylabel('Average Intensity')

plt.tight_layout()
plt.show()
```
